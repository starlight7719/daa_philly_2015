## DataPhilly events dataset: graph the ratings

```python
events_df = pd.read_pickle('events.pkl')
events_df = events_df.sort_values(by='time')
events_df
```

```python
events_df.head()
```

```python
events_df.tail(3)
```

Let's clean up the rating column. And look at some basic statistics of the result:

```python
events_df['rating_average'] = events_df['rating'].apply(lambda entry: entry['average'])
events_df['rating_average']
```

```python
events_df['rating_count'] = events_df['rating'].apply(lambda entry: entry['count'])
events_df['rating_count']
```

```python
rating_count = events_df['rating_count']
rating_average = events_df['rating_average']
```

```python
rating_count.describe()
```

```python
rating_average.describe()
```

Let's plot these with the attendance counts to explore if there is a relation.

```python
import datetime
def get_datetime_from_epoch(epoch):
    return datetime.datetime.fromtimestamp(epoch/1000.0)
```

```python
events_df['time'] = events_df['time'].apply(get_datetime_from_epoch)
events_df['time']
```

Next let's make the ``time`` column the index of the ``DataFrame`` using the ``set_index`` method and then re-``plot`` our data.

```python
events_df.set_index('time', inplace=True)
events_df['rating_count'].plot()
```

```python
all_rsvps = events_df[['yes_rsvp_count', 'waitlist_count', 'rating_count', 'rating_average']]
all_rsvps.plot(title='Ratings and Attendance over time')
```
